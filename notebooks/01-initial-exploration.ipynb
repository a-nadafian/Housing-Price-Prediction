```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style('darkgrid')
```

```python
input_data_dir = os.path.join("..", "data", "raw")
df = pd.read_csv(os.path.join(input_data_dir, "Housing.csv"))
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.histplot(data=df, x="price", kde=True)

plt.title("distribution of Housing Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=df, x="area", kde=True)

plt.title("distribution of Housing Area")
plt.xlabel("Area")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(data=df, x="furnishingstatus")

plt.title("count of Houses by Furnishing Status")
plt.xlabel("Furnishing Status")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2,2,1)
sns.countplot(data=df, x="bedrooms")

plt.title("count of Houses by Bedroom Number")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.countplot(data=df, x="bathrooms")

plt.title("count of Houses by bathrooms Number")
plt.xlabel("Number of bathrooms")
plt.ylabel("Count")

plt.subplot(2,2,3)
sns.countplot(data=df, x="stories")

plt.title("count of Houses by stories Number")
plt.xlabel("Number of stories")
plt.ylabel("Count")

plt.subplot(2,2,4)
sns.countplot(data=df, x="parking")

plt.title("count of Houses by parking Number")
plt.xlabel("Number of parking")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=df, x="area", y="price")

plt.title("Relationship between Area and Price")
plt.xlabel("Area of the House")
plt.ylabel("Price of the House")
```

```python
sns.boxenplot(data=df, x="hotwaterheating", y="price")

plt.title('Price Distribution by Air Conditioning Status')
plt.xlabel('Has Air Conditioning?')
plt.ylabel('Price')
```

```python
numeric_df = df[['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']]
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
```
